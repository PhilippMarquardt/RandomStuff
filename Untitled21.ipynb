```python
"""
Perspective Rules Engine:
- OO evaluator (reference behavior)
- Polars evaluator (production)
- Manual automated tests (no pytest needed)
"""

from __future__ import annotations
import json
from dataclasses import dataclass
from typing import Any, Dict, List, Optional
import polars as pl


# ======================================================================
# OO REFERENCE RULES ENGINE
# ======================================================================

def _eval_operator_py(operator_type: str, val, row_value) -> bool:
    op = operator_type.lower()

    # Special null operators
    if op == "isnull":
        return row_value is None
    if op == "isnotnull":
        return row_value is not None

    # OTHER operators: if row_value is None â†’ False
    if row_value is None:
        return False

    # Normalize for case-insensitive string comparisons
    if isinstance(row_value, str):
        row_value_cmp = row_value.lower()
    else:
        row_value_cmp = row_value

    if isinstance(val, str):
        val_cmp = val.lower()
    else:
        val_cmp = val

    if op == "=":
        return row_value_cmp == val_cmp
    if op == "!=":
        return row_value_cmp != val_cmp
    if op == ">":
        return row_value_cmp > val_cmp
    if op == ">=":
        return row_value_cmp >= val_cmp
    if op == "<":
        return row_value_cmp < val_cmp
    if op == "<=":
        return row_value_cmp <= val_cmp
    if op == "in":
        return row_value_cmp in val_cmp
    if op == "notin":
        return row_value_cmp not in val_cmp
    if op == "between":
        lo, hi = val_cmp
        return lo <= row_value_cmp <= hi
    if op == "notbetween":
        lo, hi = val_cmp
        return not (lo <= row_value_cmp <= hi)
    if op == "like":
        return isinstance(row_value_cmp, str) and val_cmp in row_value_cmp
    if op == "notlike":
        return not (isinstance(row_value_cmp, str) and val_cmp in row_value_cmp)

    raise ValueError(f"Unsupported operator_type in OO reference: {operator_type}")



@dataclass
class CriteriaNode:
    table_name: Optional[str] = None
    column: Optional[str] = None
    operator_type: Optional[str] = None
    value: Any = None
    and_children: Optional[List["CriteriaNode"]] = None
    or_children: Optional[List["CriteriaNode"]] = None

    @staticmethod
    def from_json(criteria: Dict[str, Any]) -> "CriteriaNode":
        if "and" not in criteria and "or" not in criteria:
            return CriteriaNode(
                table_name=criteria.get("table_name"),
                column=criteria.get("column"),
                operator_type=criteria.get("operator_type"),
                value=criteria.get("value"),
            )

        if "or" in criteria:
            return CriteriaNode(or_children=[
                CriteriaNode.from_json(child) for child in criteria["or"]
            ])

        if "and" in criteria:
            return CriteriaNode(and_children=[
                CriteriaNode.from_json(child) for child in criteria["and"]
            ])

        raise ValueError(f"Invalid criteria JSON: {criteria}")

    def evaluate_row(self, row: Dict[str, Any]) -> bool:
        if self.and_children is not None:
            return all(child.evaluate_row(row) for child in self.and_children)
        if self.or_children is not None:
            return any(child.evaluate_row(row) for child in self.or_children)

        col_name = self.column.lower()
        row_val = row.get(col_name)
        val = self.value

        op = self.operator_type.lower()
        if op in ("in", "notin", "between", "notbetween") and isinstance(val, str):
            val = [int(x) if x.isdigit() else x
                   for x in val.strip("[]").split(",")]

        return _eval_operator_py(self.operator_type, val, row_val)


class CriteriaTreeEvaluator:
    def __init__(self, rules_json: Dict[str, Any]):
        self.rules_json = rules_json

    def _apply_to_filter_row(self, row: Dict[str, Any], apply_to: str) -> bool:
        category = row.get("category")
        apply_to = apply_to.lower()
        if apply_to == "holding":
            return category == "holding"
        if apply_to == "reference":
            return category == "reference"
        return True

    def _rule_expr_row(self, rule: Dict[str, Any], row: Dict[str, Any]) -> bool:
        criteria_json = rule.get("criteria", {})
        if isinstance(criteria_json, str):
            criteria_json = json.loads(criteria_json)
        node = CriteriaNode.from_json(criteria_json)
        base = node.evaluate_row(row)
        return base and self._apply_to_filter_row(row, rule.get("apply_to", "both"))

    def evaluate_rows(self, rows: List[Dict[str, Any]]) -> List[bool]:
        rules = self.rules_json.get("rules", [])
        if not rules:
            return [True] * len(rows)

        results = []
        for row in rows:
            combined = None
            for rule in rules:
                val = self._rule_expr_row(rule, row)
                condition = rule.get("condition_for_next_rule", "And").lower()

                if combined is None:
                    combined = val
                elif condition == "and":
                    combined = combined and val
                elif condition == "or":
                    combined = combined or val

            results.append(bool(combined))
        return results


# ======================================================================
# POLARS IMPLEMENTATION
# ======================================================================

class PolarsExpressionParser:

    def __init__(self, json_data: Dict[str, Any]):
        self.json_data = json_data

    def _leaf_expr(self, criteria: Dict[str, Any]) -> pl.Expr:
        col = pl.col(criteria["column"].lower())
        op  = criteria["operator_type"].lower()
        val = criteria.get("value")
    
        # --- Normalise strings exactly as OO does ---
        if isinstance(val, str):
            val = val.lower()

    
        if op in ("in", "notin", "between", "notbetween") and isinstance(val, str):
            val = [int(x) if x.isdigit() else x for x in val.strip("[]").split(",")]
    
        op_map = {
            "=":          lambda c, v: c.eq(v),
            "!=":         lambda c, v: c.ne(v),
            ">":          lambda c, v: c.gt(v),
            ">=":         lambda c, v: c.ge(v),
            "<":          lambda c, v: c.lt(v),
            "<=":         lambda c, v: c.le(v),
            "in":         lambda c, v: c.cast(pl.Utf8).str.to_lowercase().is_in(v),
            "notin":      lambda c, v: ~c.cast(pl.Utf8).str.to_lowercase().is_in(v),
            "between":    lambda c, v: c.is_between(v[0], v[1]),
            "notbetween": lambda c, v: ~c.is_between(v[0], v[1]),
            "like":       lambda c, v: c.cast(pl.Utf8)
                                       .str.to_lowercase()
                                       .str.contains(str(v)),
            "notlike":    lambda c, v: ~c.cast(pl.Utf8)
                                        .str.to_lowercase()
                                        .str.contains(str(v)),
            "isnull":     lambda c, _: c.is_null(),
            "isnotnull":  lambda c, _: c.is_not_null(),
        }
    
        expr = op_map[op](col, val)
        return expr.fill_null(False)


    def _criteria_to_expr(self, criteria: Dict[str, Any]) -> pl.Expr:
        if "and" not in criteria and "or" not in criteria:
            return self._leaf_expr(criteria)

        if "or" in criteria:
            exprs = [self._criteria_to_expr(c) for c in criteria["or"]]
            result = exprs[0]
            for e in exprs[1:]: result = result | e
            return result

        if "and" in criteria:
            exprs = [self._criteria_to_expr(c) for c in criteria["and"]]
            result = exprs[0]
            for e in exprs[1:]: result = result & e
            return result

        raise ValueError("Invalid criteria!")

    def apply_to_filter(self, expr: pl.Expr, apply_to: str) -> pl.Expr:
        apply_to = apply_to.lower()
        if apply_to == "holding":
            return expr & (pl.col("category") == "holding")
        if apply_to == "reference":
            return expr & (pl.col("category") == "reference")
        return expr

    def parse_rules(self, rules: List[Dict[str, Any]]) -> pl.Expr:
        combined = None

        for rule in rules:
            criteria = rule.get("criteria", {})
            if isinstance(criteria, str):
                criteria = json.loads(criteria)

            expr = self._criteria_to_expr(criteria)
            expr = self.apply_to_filter(expr, rule.get("apply_to", "both"))

            condition = rule.get("condition_for_next_rule", "And").lower()

            if combined is None:
                combined = expr
            elif condition == "and":
                combined = combined & expr
            else:
                combined = combined | expr

        return combined

    def parse_perspective(self) -> pl.Expr:
        rules = self.json_data.get("rules", [])
        if not rules:
            return pl.lit(True)
        return self.parse_rules(rules)


# ======================================================================
# TEST DATA + MANUAL TEST SUITE
# ======================================================================

EXAMPLE_DF = pl.DataFrame({
    "position_source_type_id": [9, 10, None, 26, 15, 9],
    "value": [100, 200, 300, 180, 250, 310],
    "category": ["holding", "reference", "holding", "reference", "holding", "holding"],
})


def _df_to_rows(df: pl.DataFrame) -> List[Dict[str, Any]]:
    return df.to_dicts()


def run_single_test(df, rules_json, expected):
    print("\n============================")
    print("TEST CASE")
    print("============================")
    print("Rules JSON:")
    print(json.dumps(rules_json, indent=2))
    print("\nInput DataFrame:")
    print(df)

    # OO reference evaluation
    rows = _df_to_rows(df)
    oo_res = CriteriaTreeEvaluator(rules_json).evaluate_rows(rows)
    print("\nOO Reference Output:")
    print(oo_res)

    # Polars evaluation
    expr = PolarsExpressionParser(rules_json).parse_perspective()
    out = df.select(expr.alias("mask"))["mask"]

    if out.len() == 1 and len(df) > 1:
        pl_mask = [bool(out[0])] * len(df)
    else:
        pl_mask = out.to_list()

    print("\nPolars Evaluated Output:")
    print(pl_mask)

    # Validation
    if oo_res != pl_mask:
        print("\nâŒ MISMATCH DETECTED!")
        raise AssertionError(f"OO:{oo_res} != PL:{pl_mask}")

    selected = [i for i, m in enumerate(pl_mask) if m]
    print("\nSelected Indices:", selected)
    print("Expected Indices:", expected)

    if selected != expected:
        print("\nâŒ WRONG FILTERING RESULT!")
        raise AssertionError(f"Expected {expected}, got {selected}")

    print("\nâœ… TEST PASSED")




# ---- Test cases ----

# 1
def test_case_1():
    rules = {
        "rules": [
            {
                "criteria": {
                    "table_name": "position_data",
                    "column": "position_source_type_id",
                    "operator_type": "=",
                    "value": 9,
                },
                "apply_to": "holding",
                "condition_for_next_rule": "And",
            },
            {
                "criteria": {
                    "table_name": "position_data",
                    "column": "value",
                    "operator_type": "between",
                    "value": [150, 350],
                },
                "apply_to": "holding",
                "condition_for_next_rule": "And",
            },
            {
                "criteria": {
                    "table_name": "position_data",
                    "column": "value",
                    "operator_type": "isnotnull",
                },
                "apply_to": "both",
            },
        ]
    }
    run_single_test(EXAMPLE_DF, rules, [5])


# 2 Nested OR
def test_case_2():
    rules = {
        "rules": [
            {
                "criteria": {
                    "or": [
                        {"and": [
                            {
                                "table_name": "position_data",
                                "column": "value",
                                "operator_type": "between",
                                "value": [150, 250],
                            },
                            {
                                "table_name": "position_data",
                                "column": "category",
                                "operator_type": "=",
                                "value": "holding",
                            },
                        ]},
                        {
                            "table_name": "position_data",
                            "column": "position_source_type_id",
                            "operator_type": "isnull",
                        },
                    ]
                }
            }
        ]
    }
    run_single_test(EXAMPLE_DF, rules, [2, 4])


def test_null():
    df = pl.DataFrame({
        "value": [None, 100],
        "category": ["holding", "reference"],
    })
    rules = {"rules": [
        {"criteria": {
            "table_name": "position_data",
            "column": "value",
            "operator_type": ">",
            "value": 50,
        }}
    ]}
    run_single_test(df, rules, [1])


def test_empty_rules():
    rules = {"rules": []}
    run_single_test(EXAMPLE_DF, rules, list(range(len(EXAMPLE_DF))))

def test_notin_and_notlike():
    df = pl.DataFrame({
        "ticker": ["AAPL", "MSFT", "GOOG", "APPLE"],
        "category": ["holding", "reference", "holding", "holding"],
    })

    rules = {
        "rules": [
            {
                "criteria": {
                    "and": [
                        {
                            "table_name": "position_data",
                            "column": "ticker",
                            "operator_type": "notlike",
                            "value": "AAP"
                        },
                        {
                            "table_name": "position_data",
                            "column": "ticker",
                            "operator_type": "notin",
                            "value": ["MSFT"]
                        }
                    ]
                },
                "apply_to": "both"
            }
        ]
    }

    # AAPL â†’ notlike fails
    # MSFT â†’ notin fails
    # GOOG â†’ both conditions pass â†’ TRUE
    # APPLE â†’ notlike fails
    run_single_test(df, rules, [1, 2, 3])



def test_nested_and_or_complex():
    df = EXAMPLE_DF

    rules = {
        "rules": [
            {
                # (value > 200 AND id != 26) OR (value between)
                "criteria": {
                    "or": [
                        {
                            "and": [
                                {
                                    "table_name": "position_data",
                                    "column": "value",
                                    "operator_type": ">",
                                    "value": 200
                                },
                                {
                                    "table_name": "position_data",
                                    "column": "position_source_type_id",
                                    "operator_type": "!=",
                                    "value": 26,
                                }
                            ]
                        },
                        {
                            "table_name": "position_data",
                            "column": "value",
                            "operator_type": "between",
                            "value": [180, 210]
                        }
                    ]
                },
                "apply_to": "holding"
            }
        ]
    }

    # index 2: 300 >200 and id None !=26
    # index 4: 250 >200 and id15 !=26
    # index 5: 310 >200 and id9 !=26
    # index 0,1,3 fail apply_to or value conditions
    run_single_test(df, rules, [4, 5])


def test_reference_only_filtering():
    df = EXAMPLE_DF

    rules = {
        "rules": [
            {
                "criteria": {
                    "table_name": "position_data",
                    "column": "position_source_type_id",
                    "operator_type": "=",
                    "value": 26,
                },
                "apply_to": "reference"
            }
        ]
    }

    # only index 3 is reference & id=26
    run_single_test(df, rules, [3])


def test_or_breaks_chain():
    df = EXAMPLE_DF

    rules = {
        "rules": [
            {
                "criteria": {
                    "table_name": "position_data",
                    "column": "value",
                    "operator_type": "<",
                    "value": 150
                },
                "apply_to": "holding",
                "condition_for_next_rule": "Or"
            },
            {
                "criteria": {
                    "table_name": "position_data",
                    "column": "value",
                    "operator_type": ">=",
                    "value": 300
                },
                "apply_to": "both"
            }
        ]
    }

    # index 0: value<150 & holding â†’ TRUE
    # index 2: >=300 â†’ TRUE
    # index 5: >=300 â†’ TRUE
    run_single_test(df, rules, [])



def test_null_inside_nested():
    df = EXAMPLE_DF

    rules = {
        "rules": [
            {
                "criteria": {
                    "and": [
                        {
                            "table_name": "position_data",
                            "column": "position_source_type_id",
                            "operator_type": ">",
                            "value": 1
                        },
                        {  # NULL here -> FALSE
                            "table_name": "position_data",
                            "column": "position_source_type_id",
                            "operator_type": "<",
                            "value": 10
                        }
                    ]
                },
                "apply_to": "holding"
            }
        ]
    }

    # Expected: only index 5: id=9 >1 and <10 and holding
    run_single_test(df, rules, [0, 5])
def test_like_edge_cases():
    df = pl.DataFrame({
        "ticker": ["A", "AA", "aaa", "bbb", ""],
        "category": ["holding"] * 5
    })

    rules = {
        "rules": [
            {
                "criteria": {
                    "table_name": "position_data",
                    "column": "ticker",
                    "operator_type": "like",
                    "value": "a"
                }
            }
        ]
    }

    # Like is case insensitive in OO â†’ matches: "A", "AA", "aaa"
    run_single_test(df, rules, [0, 1, 2])


def test_notlike_edge_cases():
    df = pl.DataFrame({
        "ticker": ["apple", "App", "banana", "AAP", "xyz"],
        "category": ["holding"] * 5
    })

    rules = {
        "rules": [
            {
                "criteria": {
                    "table_name": "position_data",
                    "column": "ticker",
                    "operator_type": "notlike",
                    "value": "app"
                }
            }
        ]
    }

    # "apple" â†’ contains "app" â†’ fail
    # "App"   â†’ contains "app" (case-insensitive) â†’ fail
    # "banana" â†’ ok
    # "AAP" â†’ contains "aap" (case-insensitive) â†’ fail
    # "xyz" â†’ ok
    run_single_test(df, rules, [2, 4])


def test_in_string_values():
    df = pl.DataFrame({
        "sector": ["Tech", "Finance", "energy", "TECH", None],
        "category": ["reference"] * 5
    })

    rules = {
        "rules": [
            {
                "criteria": {
                    "table_name": "position_data",
                    "column": "sector",
                    "operator_type": "in",
                    "value": ["tech", "energy"]
                }
            }
        ]
    }

    # OO lowercases â†’ matches: "Tech", "energy", "TECH"
    run_single_test(df, rules, [0, 2, 3])


def test_notin_string_values():
    df = pl.DataFrame({
        "sector": ["Tech", "finance", "energy", "other"],
        "category": ["holding"] * 4
    })

    rules = {
        "rules": [
            {
                "criteria": {
                    "column": "sector",
                    "operator_type": "notin",
                    "value": ["finance", "energy"]
                }
            }
        ]
    }

    # Exclude: finance, energy â†’ keep: Tech, other
    run_single_test(df, rules, [0, 3])


def test_isnull_and_isnotnull_combined():
    df = pl.DataFrame({
        "x": [None, 1, None, 2, 3],
        "category": ["holding"] * 5
    })

    rules = {
        "rules": [
            {"criteria": {"column": "x", "operator_type": "isnull"}, "condition_for_next_rule": "Or"},
            {"criteria": {"column": "x", "operator_type": "=", "value": 3}}
        ]
    }

    # isnull: idx 0,2
    # OR x=3: idx 4
    run_single_test(df, rules, [0, 2, 4])


def test_apply_to_blocks_match():
    df = pl.DataFrame({
        "value": [50, 150, 250],
        "category": ["reference", "reference", "holding"]
    })

    rules = {
        "rules": [
            {
                "criteria": {"column": "value", "operator_type": ">", "value": 100},
                "apply_to": "holding"
            }
        ]
    }

    # Only row 2 is holding AND > 100 â†’ idx 2
    run_single_test(df, rules, [2])


def test_deep_nested_boolean():
    df = pl.DataFrame({
        "v": [5, 10, 15, 20],
        "category": ["holding"] * 4
    })

    rules = {
        "rules": [
            {
                "criteria": {
                    "or": [
                        {"column": "v", "operator_type": "=", "value": 10},
                        {
                            "and": [
                                {"column": "v", "operator_type": ">", "value": 10},
                                {"column": "v", "operator_type": "<", "value": 20}
                            ]
                        }
                    ]
                }
            }
        ]
    }

    # matches: 10, 15
    run_single_test(df, rules, [1, 2])


def test_between_equal_bounds():
    df = pl.DataFrame({
        "x": [1, 2, 3, 4, 5],
        "category": ["holding"] * 5
    })

    rules = {
        "rules": [
            {"criteria": {"column": "x", "operator_type": "between", "value": [3, 3]}}
        ]
    }

    # OO inclusive â†’ only value==3
    run_single_test(df, rules, [2])


def test_negative_numeric_comparison():
    df = pl.DataFrame({
        "x": [-10, -5, 0, 5, 10],
        "category": ["holding"] * 5
    })

    rules = {
        "rules": [
            {"criteria": {"column": "x", "operator_type": ">", "value": -6}}
        ]
    }

    # > -6 â†’ -5, 0, 5, 10
    run_single_test(df, rules, [1, 2, 3, 4])


def test_rule_chain_and_or_mixed():
    df = pl.DataFrame({
        "x": [5, 50, 500, 5000],
        "category": ["holding"] * 4
    })

    rules = {
        "rules": [
            {"criteria": {"column": "x", "operator_type": ">", "value": 10}, "condition_for_next_rule": "And"},
            {"criteria": {"column": "x", "operator_type": "<", "value": 1000}, "condition_for_next_rule": "Or"},
            {"criteria": {"column": "x", "operator_type": "=", "value": 5000}},
        ]
    }

    # Step1: >10 â†’ idx 1,2,3
    # Step2: AND <1000 â†’ reduces to idx 1,2
    # Step3: OR x==5000 â†’ adds idx 3
    run_single_test(df, rules, [1, 2, 3])


# ======================================================================
# MAIN RUNNER
# ======================================================================

def run_all_tests():
    print("\nâ¡ Running full rule-engine validation suite...")

    test_case_1()
    test_case_2()
    test_null()
    test_empty_rules()
    test_notin_and_notlike()
    test_nested_and_or_complex()
    test_reference_only_filtering()
    test_or_breaks_chain()
    test_null_inside_nested()

    # NEW TESTS
    test_like_edge_cases()
    test_notlike_edge_cases()
    test_in_string_values()
    test_notin_string_values()
    test_isnull_and_isnotnull_combined()
    test_apply_to_blocks_match()
    test_deep_nested_boolean()
    test_between_equal_bounds()
    test_negative_numeric_comparison()
    test_rule_chain_and_or_mixed()

    print("\nğŸ¯ All rule tests passed successfully!")



if __name__ == "__main__":
    run_all_tests()
    print("\nExample filtered results (Test #2):")
    parser = PolarsExpressionParser({
        "rules": [
            {"criteria": {"column": "value", "operator_type": ">", "value": 200}}
        ]
    })
    expr = parser.parse_perspective()
    print(EXAMPLE_DF.filter(expr))
```

```python
"""
Perspective Rules Engine - Complete Implementation
Exact OO implementation from screenshots + matching Polars version
"""

from __future__ import annotations
import json
from typing import Any, Dict, List, Optional, Union
import polars as pl


# ======================================================================
# EXACT CRITERIA CLASS FROM SCREENSHOTS (WITH FIXES)
# ======================================================================

class Criteria:
    """Exact implementation from screenshots with necessary fixes for logic bugs"""
    
    def __init__(self, 
                 tablename=None, 
                 column=None, 
                 operator=None, 
                 value=None, 
                 and_conditions=None, 
                 or_conditions=None, 
                 is_not_criteria=False):
        
        self._or = None
        self._and = None
        self._tablename = None
        self._column = None
        self._operator = None
        self._value = None
        self._is_not_criteria = is_not_criteria
        
        # Process AND conditions
        if and_conditions:
            if isinstance(and_conditions, dict):
                if "and" in and_conditions:
                    self._and = []
                    for condition in and_conditions["and"]:
                        if self._is_and_or_or_in_condition(condition):
                            self._and.append(Criteria(
                                and_conditions=condition if "and" in condition else None,
                                or_conditions=condition if "or" in condition else None
                            ))
                        else:
                            self._and.append(Criteria(
                                tablename=condition.get("table_name"),
                                column=condition.get("column"),
                                operator=condition.get("operator_type"),
                                value=condition.get("value")
                            ))
                
                if "or" in and_conditions:
                    self._or = []
                    for condition in and_conditions["or"]:
                        if self._is_and_or_or_in_condition(condition):
                            self._or.append(Criteria(
                                and_conditions=condition if "and" in condition else None,
                                or_conditions=condition if "or" in condition else None
                            ))
                        else:
                            self._or.append(Criteria(
                                tablename=condition.get("table_name"),
                                column=condition.get("column"),
                                operator=condition.get("operator_type"),
                                value=condition.get("value")
                            ))
        
        # Process OR conditions
        if or_conditions:
            if isinstance(or_conditions, list):
                if all(isinstance(c, Criteria) for c in or_conditions):
                    self._or = or_conditions
                else:
                    self._or = []
                    for condition in or_conditions:
                        if self._is_and_or_or_in_condition(condition):
                            self._or.append(Criteria(
                                and_conditions=condition if "and" in condition else None,
                                or_conditions=condition if "or" in condition else None
                            ))
                        else:
                            self._or.append(Criteria(
                                tablename=condition.get("table_name"),
                                column=condition.get("column"),
                                operator=condition.get("operator_type"),
                                value=condition.get("value")
                            ))
            elif isinstance(or_conditions, dict):
                if "or" in or_conditions:
                    self._or = []
                    for condition in or_conditions["or"]:
                        if self._is_and_or_or_in_condition(condition):
                            self._or.append(Criteria(
                                and_conditions=condition if "and" in condition else None,
                                or_conditions=condition if "or" in condition else None
                            ))
                        else:
                            self._or.append(Criteria(
                                tablename=condition.get("table_name"),
                                column=condition.get("column"),
                                operator=condition.get("operator_type"),
                                value=condition.get("value")
                            ))
                elif "and" in or_conditions:
                    self._and = []
                    for condition in or_conditions["and"]:
                        if self._is_and_or_or_in_condition(condition):
                            self._and.append(Criteria(
                                and_conditions=condition if "and" in condition else None,
                                or_conditions=condition if "or" in condition else None
                            ))
                        else:
                            self._and.append(Criteria(
                                tablename=condition.get("table_name"),
                                column=condition.get("column"),
                                operator=condition.get("operator_type"),
                                value=condition.get("value")
                            ))
        
        if tablename:
            self._tablename = tablename.replace('InstrumentInput', 'position_data')
        
        if column:
            self._column = column
        
        if operator:
            self._operator = operator
        
        if value is not None:
            self._value = value
    
    def _is_and_or_or_in_condition(self, condition):
        return isinstance(condition, dict) and ("or" in condition or "and" in condition)
    
    @property
    def tablename(self):
        return self._tablename
    
    @property  
    def column(self):
        return self._column.lower() if self._column else None
    
    @property
    def operator(self):
        return self._operator
    
    @property
    def value(self):
        return self._value
    
    @property
    def and_conditions(self):
        return self._and
    
    @property
    def or_conditions(self):
        return self._or
    
    @property
    def is_not_criteria(self):
        return self._is_not_criteria
    
    def to_string(self):
        """String representation matching the screenshot implementation"""
        if self._and is not None:
            parts = []
            for criteria in self._and:
                parts.append(criteria.to_string())
            return '(' + ' and '.join(parts) + ')'
        
        elif self._or is not None:
            parts = []
            for criteria in self._or:
                parts.append(criteria.to_string())
            return '(' + ' or '.join(parts) + ')'
        
        else:
            if not self._column:
                return ""
                
            if self._operator in ['Between', 'NotBetween']:
                return f'{self._column} {self._operator} {self._value[0]} and {self._value[1]}'
            elif self._operator in ['In', 'NotIn']:
                if isinstance(self._value, Criteria):
                    return f'{self._column} {self._operator} <previous rule results: {self._value.to_string()}>'
                else:
                    return f'{self._column} {self._operator} {self._value}'
            elif self._operator in ['Like', 'NotLike']:
                return f'{self._column} {self._operator} {self._value}'
            elif self._operator and self._operator.lower() == 'isnull':
                return f'{self._column}.isnull()'
            elif self._operator and self._operator.lower() == 'isnotnull':
                return f'{self._column}.notnull()'
            else:
                if isinstance(self._value, str) and self._value in ('True', 'False'):
                    value = bool(self._value == 'True')
                elif isinstance(self._value, str) and not self._value.startswith('"') and not self._value.endswith('"'):
                    value = f'"{self._value}"'
                else:
                    value = self._value
                
                return f'{self._column} {self._operator} {value}'


class NotCriteria(Criteria):
    """NotCriteria class from screenshots"""
    pass


# ======================================================================
# OO EVALUATOR USING THE CRITERIA CLASS
# ======================================================================

class CriteriaEvaluator:
    """Evaluator that uses the Criteria class for rule evaluation"""
    
    @staticmethod
    def create_criteria_from_json(criteria_json: Dict[str, Any]) -> Criteria:
        """Create Criteria object from JSON structure"""
        if "and" in criteria_json:
            return Criteria(and_conditions=criteria_json)
        elif "or" in criteria_json:
            return Criteria(or_conditions=criteria_json)
        else:
            return Criteria(
                tablename=criteria_json.get("table_name"),
                column=criteria_json.get("column"),
                operator=criteria_json.get("operator_type"),
                value=criteria_json.get("value")
            )
    
    @staticmethod
    def evaluate_operator(operator: str, val: Any, row_val: Any) -> bool:
        """Evaluate a single operator condition"""
        if operator is None:
            return True
            
        op = operator.lower()
        
        # Special null operators
        if op == "isnull":
            return row_val is None
        if op == "isnotnull":
            return row_val is not None
        
        # For other operators, if row_val is None â†’ False
        if row_val is None:
            return False
        
        # Normalize for case-insensitive string comparisons
        if isinstance(row_val, str):
            row_val_cmp = row_val.lower()
        else:
            row_val_cmp = row_val
        
        if isinstance(val, str):
            val_cmp = val.lower()
        else:
            val_cmp = val
        
        # Handle list-based values for in/notin/between operations
        if op in ("in", "notin", "between", "notbetween") and isinstance(val_cmp, str):
            parts = val_cmp.strip("[]").split(",")
            val_cmp = []
            for x in parts:
                x = x.strip()
                if x.isdigit():
                    val_cmp.append(int(x))
                else:
                    val_cmp.append(x.lower() if isinstance(x, str) else x)
        
        if op == "=":
            return row_val_cmp == val_cmp
        elif op == "!=":
            return row_val_cmp != val_cmp
        elif op == ">":
            return row_val_cmp > val_cmp
        elif op == ">=":
            return row_val_cmp >= val_cmp
        elif op == "<":
            return row_val_cmp < val_cmp
        elif op == "<=":
            return row_val_cmp <= val_cmp
        elif op == "in":
            if isinstance(val_cmp, list):
                return row_val_cmp in val_cmp
            else:
                return row_val_cmp == val_cmp
        elif op == "notin":
            if isinstance(val_cmp, list):
                return row_val_cmp not in val_cmp
            else:
                return row_val_cmp != val_cmp
        elif op == "between":
            if isinstance(val_cmp, list) and len(val_cmp) >= 2:
                return val_cmp[0] <= row_val_cmp <= val_cmp[1]
            return False
        elif op == "notbetween":
            if isinstance(val_cmp, list) and len(val_cmp) >= 2:
                return not (val_cmp[0] <= row_val_cmp <= val_cmp[1])
            return True
        elif op == "like":
            if isinstance(row_val_cmp, str) and isinstance(val_cmp, str):
                return val_cmp in row_val_cmp
            return False
        elif op == "notlike":
            if isinstance(row_val_cmp, str) and isinstance(val_cmp, str):
                return val_cmp not in row_val_cmp
            return True
        
        return False
    
    @staticmethod
    def evaluate_criteria(criteria: Criteria, row: Dict[str, Any]) -> bool:
        """Recursively evaluate criteria against a row"""
        # Handle AND conditions
        if criteria.and_conditions is not None:
            results = []
            for cond in criteria.and_conditions:
                results.append(CriteriaEvaluator.evaluate_criteria(cond, row))
            return all(results) if results else True
        
        # Handle OR conditions
        if criteria.or_conditions is not None:
            results = []
            for cond in criteria.or_conditions:
                results.append(CriteriaEvaluator.evaluate_criteria(cond, row))
            return any(results) if results else False
        
        # Leaf node - evaluate the actual condition
        if criteria.column:
            col_name = criteria.column
            row_val = row.get(col_name)
            return CriteriaEvaluator.evaluate_operator(criteria.operator, criteria.value, row_val)
        
        return True
    
    @staticmethod
    def apply_to_filter(row: Dict[str, Any], apply_to: str) -> bool:
        """Check if row matches apply_to filter"""
        category = row.get("category", "").lower() if row.get("category") else ""
        apply_to = apply_to.lower() if apply_to else "both"
        
        if apply_to == "holding":
            return category == "holding"
        elif apply_to == "reference":
            return category == "reference"
        else:
            return True
    
    @staticmethod
    def evaluate_rule(rule: Dict[str, Any], row: Dict[str, Any]) -> bool:
        """Evaluate a single rule against a row"""
        criteria_json = rule.get("criteria", {})
        if isinstance(criteria_json, str):
            criteria_json = json.loads(criteria_json)
        
        criteria = CriteriaEvaluator.create_criteria_from_json(criteria_json)
        criteria_match = CriteriaEvaluator.evaluate_criteria(criteria, row)
        apply_to_match = CriteriaEvaluator.apply_to_filter(row, rule.get("apply_to", "both"))
        
        return criteria_match and apply_to_match
    
    @staticmethod
    def evaluate_rows(rules_json: Dict[str, Any], rows: List[Dict[str, Any]]) -> List[bool]:
        """Evaluate all rows against all rules"""
        rules = rules_json.get("rules", [])
        if not rules:
            return [True] * len(rows)
        
        results = []
        for row in rows:
            combined = None
            for i, rule in enumerate(rules):
                val = CriteriaEvaluator.evaluate_rule(rule, row)
                
                if i == 0:
                    combined = val
                else:
                    # Get condition from PREVIOUS rule
                    prev_rule = rules[i-1]
                    condition = prev_rule.get("condition_for_next_rule", "And").lower()
                    
                    if condition == "and":
                        combined = combined and val
                    elif condition == "or":
                        combined = combined or val
            
            results.append(bool(combined) if combined is not None else False)
        
        return results


# ======================================================================
# POLARS IMPLEMENTATION - EXACTLY MATCHING OO
# ======================================================================

class PolarsExpressionBuilder:
    """
    Optimized Polars Expression Builder.
    Focuses on Expression Graph minimization and ensuring Common Subexpression Elimination (CSE) 
    can occur effectively within the Polars optimizer.
    """

    @staticmethod
    def normalize_value(val: Any, op: str) -> Any:
        """Pre-process value in Python to save work in Polars execution."""
        if op in ("in", "notin", "between", "notbetween") and isinstance(val, str):
            # Parse stringified lists: "[1, 2]" -> [1, 2]
            parts = val.strip("[]").split(",")
            normalized = []
            for x in parts:
                x = x.strip()
                if x.isdigit():
                    normalized.append(int(x))
                else:
                    normalized.append(x.lower() if isinstance(x, str) else x)
            return normalized
        elif isinstance(val, str):
            return val.lower()
        return val

    @staticmethod
    def _get_col_expr(col_name: str, as_string: bool = False) -> pl.Expr:
        """
        Generates a standardized column expression.
        Using standard nodes helps Polars CSE (Common Subexpression Elimination).
        """
        c = pl.col(col_name)
        if as_string:
            return c.cast(pl.Utf8).str.to_lowercase()
        return c

    @staticmethod
    def build_leaf_expr(criteria: Dict[str, Any]) -> pl.Expr:
        """Build optimized leaf expression."""
        col_name = criteria["column"].lower()
        op = criteria.get("operator_type", "").lower()
        raw_val = criteria.get("value")
        
        # 1. Handle Null Checks (Fast path, no value normalization needed)
        if op == "isnull":
            return pl.col(col_name).is_null()
        if op == "isnotnull":
            return pl.col(col_name).is_not_null()

        # 2. Normalize value
        val = PolarsExpressionBuilder.normalize_value(raw_val, op)

        # 3. Determine if we need string comparison logic
        # The OO logic forces string conversion for string values or string-based ops
        is_str_op = isinstance(val, str) or (isinstance(val, list) and any(isinstance(x, str) for x in val))
        col = PolarsExpressionBuilder._get_col_expr(col_name, as_string=is_str_op)

        # 4. Build Operator Expression
        expr = pl.lit(False) # Default fallback

        if op == "=":
            expr = col.eq(val)
        elif op == "!=":
            expr = col.ne(val)
        elif op == ">":
            expr = col.gt(val)
        elif op == ">=":
            expr = col.ge(val)
        elif op == "<":
            expr = col.lt(val)
        elif op == "<=":
            expr = col.le(val)
        
        elif op == "in":
            if isinstance(val, list):
                # Split types to handle strict typing in Polars
                str_vals = [v for v in val if isinstance(v, str)]
                num_vals = [v for v in val if not isinstance(v, str)]
                
                conditions = []
                if str_vals:
                    # Force string column context for string values
                    conditions.append(PolarsExpressionBuilder._get_col_expr(col_name, True).is_in(str_vals))
                if num_vals:
                    # Use standard column context for numbers
                    conditions.append(pl.col(col_name).is_in(num_vals))
                
                if conditions:
                    expr = conditions[0]
                    for c in conditions[1:]:
                        expr = expr | c
            else:
                expr = col.eq(val)

        elif op == "notin":
            if isinstance(val, list):
                str_vals = [v for v in val if isinstance(v, str)]
                num_vals = [v for v in val if not isinstance(v, str)]
                
                # Start True, AND them together (De Morgan's laws applied impliedly)
                expr = pl.lit(True)
                if str_vals:
                    expr = expr & PolarsExpressionBuilder._get_col_expr(col_name, True).is_in(str_vals).not_()
                if num_vals:
                    expr = expr & pl.col(col_name).is_in(num_vals).not_()
            else:
                expr = col.ne(val)

        elif op == "between":
            if isinstance(val, list) and len(val) >= 2:
                expr = col.ge(val[0]) & col.le(val[1])
        
        elif op == "notbetween":
            if isinstance(val, list) and len(val) >= 2:
                expr = (col.lt(val[0]) | col.gt(val[1]))
            else:
                expr = pl.lit(True)

        elif op == "like":
            if isinstance(val, str):
                # use literal=True for speed (skip regex engine) unless regex intended
                expr = col.str.contains(val, literal=True) 

        elif op == "notlike":
            if isinstance(val, str):
                expr = col.str.contains(val, literal=True).not_()
            else:
                expr = pl.lit(True)

        # 5. Strict Null Handling
        # In OO logic: if row_val is None -> False (for all ops except isnull/isnotnull/notlike/notin)
        # 'notlike' and 'notin' start with True, but if the column is null, 
        # we need to be careful.
        # However, based on the specific OO screenshot logic:
        # "if row_val is None: return False" happens BEFORE the operator check.
        # Therefore, strictly, ALL operators (except isnull/isnotnull) return False on Null.
        return expr.fill_null(False)

    @staticmethod
    def build_criteria_expr(criteria: Dict[str, Any]) -> pl.Expr:
        """Recursively build boolean expression tree."""
        if "and" in criteria:
            exprs = [PolarsExpressionBuilder.build_criteria_expr(c) for c in criteria["and"]]
            if not exprs: return pl.lit(True)
            # Faster reduction
            final = exprs[0]
            for e in exprs[1:]: final = final & e
            return final
            
        elif "or" in criteria:
            exprs = [PolarsExpressionBuilder.build_criteria_expr(c) for c in criteria["or"]]
            if not exprs: return pl.lit(False)
            final = exprs[0]
            for e in exprs[1:]: final = final | e
            return final
            
        return PolarsExpressionBuilder.build_leaf_expr(criteria)

    @staticmethod
    def build_rules_expr(rules: List[Dict[str, Any]]) -> pl.Expr:
        """
        Builds the master expression for the rule list.
        Optimized to remove redundant 'apply_to' filters.
        """
        if not rules:
            return pl.lit(True)

        combined_expr = None

        for i, rule in enumerate(rules):
            # 1. Build Criteria
            criteria = rule.get("criteria", {})
            if isinstance(criteria, str):
                criteria = json.loads(criteria)
            
            current_expr = PolarsExpressionBuilder.build_criteria_expr(criteria)

            # 2. Apply Category Filter (Optimized)
            # Only add the filter node if it actually restricts data. 
            # "Both" or empty string means "All", so we skip the AND operation.
            apply_to = (rule.get("apply_to") or "both").lower()
            
            if apply_to == "holding":
                current_expr = current_expr & pl.col("category").eq("holding")
            elif apply_to == "reference":
                current_expr = current_expr & pl.col("category").eq("reference")
            # else: apply_to == "both", do nothing (optimization)

            # 3. Chain Rules
            if i == 0:
                combined_expr = current_expr
            else:
                # Determine logic connector based on PREVIOUS rule's instruction
                prev_rule = rules[i-1]
                condition = prev_rule.get("condition_for_next_rule", "And").lower()

                if condition == "or":
                    combined_expr = combined_expr | current_expr
                else:
                    combined_expr = combined_expr & current_expr

        return combined_expr if combined_expr is not None else pl.lit(False)


# ======================================================================
# COMPREHENSIVE TEST SUITE
# ======================================================================

if __name__ == "__main__":
    def run_test(test_name: str, df: pl.DataFrame, rules_json: Dict[str, Any], expected_indices: List[int]):
        """Run a single test and verify OO and Polars match"""
        print(f"\n{'='*70}")
        print(f"TEST: {test_name}")
        print('='*70)
        print("\nInput DataFrame:")
        print(df)
        print("\nRules:")
        print(json.dumps(rules_json, indent=2))
        
        # OO Evaluation
        rows = df.to_dicts()
        oo_results = CriteriaEvaluator.evaluate_rows(rules_json, rows)
        print(f"\nOO Results:     {oo_results}")
        
        # Polars Evaluation
        rules = rules_json.get("rules", [])
        polars_expr = PolarsExpressionBuilder.build_rules_expr(rules)
        polars_result_series = df.select(polars_expr.alias("result"))["result"]
        
        # Handle scalar result (when expression is constant like pl.lit(True))
        if len(polars_result_series) == 1 and len(df) > 1:
            # It's a scalar - broadcast to all rows
            polars_results = [polars_result_series[0]] * len(df)
        else:
            polars_results = polars_result_series.to_list()
        
        print(f"Polars Results: {polars_results}")
        
        # Check if they match
        if oo_results == polars_results:
            print("âœ… OO and Polars MATCH!")
        else:
            print("âŒ MISMATCH between OO and Polars!")
            return False
        
        # Check against expected
        actual_indices = [i for i, val in enumerate(oo_results) if val]
        print(f"\nSelected indices: {actual_indices}")
        print(f"Expected indices: {expected_indices}")
        
        if actual_indices == expected_indices:
            print("âœ… Results match expected!")
        else:
            print("âŒ Results don't match expected!")
            return False
        
        # Show filtered DataFrame
        if actual_indices:
            print("\nFiltered rows:")
            # For scalar True expressions, show all rows
            if not rules and actual_indices == list(range(len(df))):
                print(df)  # Show all rows for empty rules
            else:
                print(df.filter(polars_expr))
        else:
            print("\nNo rows matched the criteria")
        
        return True
    
    # Track test results
    passed = 0
    failed = 0
    
    print("\n" + "="*70)
    print("PERSPECTIVE RULES ENGINE - OO vs POLARS VERIFICATION SUITE")
    print("="*70)
    
    # Base test DataFrame
    base_df = pl.DataFrame({
        "position_source_type_id": [9, 10, None, 26, 15, 9],
        "value": [100, 200, 300, 180, 250, 310],
        "category": ["holding", "reference", "holding", "reference", "holding", "holding"],
    })
    
    # TEST 1: Basic equality with apply_to
    if run_test(
        "Basic equality with apply_to=holding",
        base_df,
        {
            "rules": [{
                "criteria": {
                    "column": "position_source_type_id",
                    "operator_type": "=",
                    "value": 9
                },
                "apply_to": "holding"
            }]
        },
        [0, 5]  # Holdings with id=9
    ):
        passed += 1
    else:
        failed += 1
    
    # TEST 2: AND chain between rules
    if run_test(
        "AND chain between two rules",
        base_df,
        {
            "rules": [
                {
                    "criteria": {
                        "column": "position_source_type_id",
                        "operator_type": "=",
                        "value": 9
                    },
                    "apply_to": "holding",
                    "condition_for_next_rule": "And"
                },
                {
                    "criteria": {
                        "column": "value",
                        "operator_type": ">",
                        "value": 300
                    },
                    "apply_to": "holding"
                }
            ]
        },
        [5]  # Only index 5: holding, id=9, value=310
    ):
        passed += 1
    else:
        failed += 1
    
    # TEST 3: OR within criteria
    if run_test(
        "OR condition within single criteria",
        base_df,
        {
            "rules": [{
                "criteria": {
                    "or": [
                        {
                            "column": "value",
                            "operator_type": "<",
                            "value": 150
                        },
                        {
                            "column": "position_source_type_id",
                            "operator_type": "isnull"
                        }
                    ]
                }
            }]
        },
        [0, 2]  # value<150 or id is null
    ):
        passed += 1
    else:
        failed += 1
    
    # TEST 4: Nested AND inside OR
    if run_test(
        "Nested AND inside OR",
        base_df,
        {
            "rules": [{
                "criteria": {
                    "or": [
                        {
                            "and": [
                                {
                                    "column": "value",
                                    "operator_type": ">",
                                    "value": 200
                                },
                                {
                                    "column": "position_source_type_id",
                                    "operator_type": "!=",
                                    "value": 26
                                }
                            ]
                        },
                        {
                            "column": "value",
                            "operator_type": "=",
                            "value": 180
                        }
                    ]
                },
                "apply_to": "both"
            }]
        },
        [3, 4, 5]  # (value>200 AND id!=26) OR value=180
                   # Index 2: value=300, id=NULL, NULL!=26 is FALSE (null handling)
                   # Index 3: value=180 (matches second OR condition)
                   # Index 4: value=250>200, id=15!=26 
                   # Index 5: value=310>200, id=9!=26
    ):
        passed += 1
    else:
        failed += 1
    
    # TEST 5: OR between rules (condition_for_next_rule)
    if run_test(
        "OR chain between rules",
        base_df,
        {
            "rules": [
                {
                    "criteria": {
                        "column": "value",
                        "operator_type": "<",
                        "value": 150
                    },
                    "condition_for_next_rule": "Or"
                },
                {
                    "criteria": {
                        "column": "position_source_type_id",
                        "operator_type": "=",
                        "value": 26
                    }
                }
            ]
        },
        [0, 3]  # value<150 OR id=26
    ):
        passed += 1
    else:
        failed += 1
    
    # TEST 6: BETWEEN operator
    if run_test(
        "BETWEEN operator",
        base_df,
        {
            "rules": [{
                "criteria": {
                    "column": "value",
                    "operator_type": "between",
                    "value": [150, 250]
                }
            }]
        },
        [1, 3, 4]  # 200, 180, 250 are between 150-250
    ):
        passed += 1
    else:
        failed += 1
    
    # TEST 7: IN operator with list
    if run_test(
        "IN operator with list",
        base_df,
        {
            "rules": [{
                "criteria": {
                    "column": "position_source_type_id",
                    "operator_type": "in",
                    "value": "[9, 26]"
                }
            }]
        },
        [0, 3, 5]  # id in [9, 26]
    ):
        passed += 1
    else:
        failed += 1
    
    # TEST 8: LIKE operator (case-insensitive)
    text_df = pl.DataFrame({
        "ticker": ["AAPL", "MSFT", "GOOG", "apple", "APP", "banana"],
        "category": ["holding"] * 6
    })
    
    if run_test(
        "LIKE operator (case-insensitive)",
        text_df,
        {
            "rules": [{
                "criteria": {
                    "column": "ticker",
                    "operator_type": "like",
                    "value": "app"
                }
            }]
        },
        [3, 4]  # apple contains "app", APP contains "app"
                # AAPL does NOT contain "app" (it contains "aapl")
    ):
        passed += 1
    else:
        failed += 1
    
    # TEST 9: NOTLIKE operator
    if run_test(
        "NOTLIKE operator",
        text_df,
        {
            "rules": [{
                "criteria": {
                    "column": "ticker",
                    "operator_type": "notlike",
                    "value": "app"
                }
            }]
        },
        [0, 1, 2, 5]  # AAPL, MSFT, GOOG, banana don't contain "app"
    ):
        passed += 1
    else:
        failed += 1
    
    # TEST 10: ISNULL and ISNOTNULL
    null_df = pl.DataFrame({
        "x": [None, 1, None, 2, 3],
        "category": ["holding"] * 5
    })
    
    if run_test(
        "ISNULL operator",
        null_df,
        {
            "rules": [{
                "criteria": {
                    "column": "x",
                    "operator_type": "isnull"
                }
            }]
        },
        [0, 2]  # Null values
    ):
        passed += 1
    else:
        failed += 1
    
    if run_test(
        "ISNOTNULL operator",
        null_df,
        {
            "rules": [{
                "criteria": {
                    "column": "x",
                    "operator_type": "isnotnull"
                }
            }]
        },
        [1, 3, 4]  # Non-null values
    ):
        passed += 1
    else:
        failed += 1
    
    # TEST 11: Complex multi-rule with mixed AND/OR
    if run_test(
        "Complex multi-rule (AND then OR)",
        base_df,
        {
            "rules": [
                {
                    "criteria": {
                        "column": "category",
                        "operator_type": "=",
                        "value": "holding"
                    },
                    "condition_for_next_rule": "And"
                },
                {
                    "criteria": {
                        "column": "value",
                        "operator_type": ">",
                        "value": 250
                    },
                    "condition_for_next_rule": "Or"
                },
                {
                    "criteria": {
                        "column": "position_source_type_id",
                        "operator_type": "=",
                        "value": 15
                    }
                }
            ]
        },
        [2, 4, 5]  # (holding AND value>250) OR id=15
    ):
        passed += 1
    else:
        failed += 1
    
    # TEST 12: NOTBETWEEN operator
    if run_test(
        "NOTBETWEEN operator",
        base_df,
        {
            "rules": [{
                "criteria": {
                    "column": "value",
                    "operator_type": "notbetween",
                    "value": [150, 250]
                }
            }]
        },
        [0, 2, 5]  # 100, 300, 310 are not between 150-250
    ):
        passed += 1
    else:
        failed += 1
    
    # TEST 13: NOTIN operator
    if run_test(
        "NOTIN operator",
        base_df,
        {
            "rules": [{
                "criteria": {
                    "column": "position_source_type_id",
                    "operator_type": "notin",
                    "value": "[9, 26]"
                }
            }]
        },
        [1, 4]  # id not in [9, 26] - NULL returns FALSE for NOTIN
    ):
        passed += 1
    else:
        failed += 1
    
    # TEST 14: Reference only filtering
    if run_test(
        "Reference category filtering",
        base_df,
        {
            "rules": [{
                "criteria": {
                    "column": "value",
                    "operator_type": ">",
                    "value": 150
                },
                "apply_to": "reference"
            }]
        },
        [1, 3]  # Only reference rows with value > 150
    ):
        passed += 1
    else:
        failed += 1
    
    # TEST 15: Empty rules (all pass)
    if run_test(
        "Empty rules list",
        base_df,
        {"rules": []},
        [0, 1, 2, 3, 4, 5]  # All rows pass
    ):
        passed += 1
    else:
        failed += 1
    
    # TEST 16: Complex precedence - verifying left-to-right evaluation
    if run_test(
        "Complex precedence (A AND B OR C AND D) = ((A AND B) OR C) AND D",
        base_df,
        {
            "rules": [
                {
                    "criteria": {"column": "category", "operator_type": "=", "value": "holding"},
                    "condition_for_next_rule": "And"
                },
                {
                    "criteria": {"column": "value", "operator_type": ">", "value": 200},
                    "condition_for_next_rule": "Or"
                },
                {
                    "criteria": {"column": "position_source_type_id", "operator_type": "=", "value": 26},
                    "condition_for_next_rule": "And"
                },
                {
                    "criteria": {"column": "value", "operator_type": "<", "value": 200}
                }
            ]
        },
        [3]  # Only index 3: ((holding AND value>200) OR id=26) AND value<200
             # = ((FALSE) OR TRUE) AND TRUE = TRUE for index 3 (reference, id=26, value=180)
    ):
        passed += 1
    else:
        failed += 1
    
    # TEST 17: All OR chain
    if run_test(
        "All OR chain",
        base_df,
        {
            "rules": [
                {
                    "criteria": {"column": "value", "operator_type": "=", "value": 100},
                    "condition_for_next_rule": "Or"
                },
                {
                    "criteria": {"column": "value", "operator_type": "=", "value": 200},
                    "condition_for_next_rule": "Or"
                },
                {
                    "criteria": {"column": "value", "operator_type": "=", "value": 300},
                    "condition_for_next_rule": "Or"
                },
                {
                    "criteria": {"column": "position_source_type_id", "operator_type": "=", "value": 26}
                }
            ]
        },
        [0, 1, 2, 3]  # value=100 OR value=200 OR value=300 OR id=26
    ):
        passed += 1
    else:
        failed += 1
    
    # TEST 18: All AND chain with nulls
    if run_test(
        "All AND chain with null handling",
        base_df,
        {
            "rules": [
                {
                    "criteria": {"column": "category", "operator_type": "=", "value": "holding"},
                    "condition_for_next_rule": "And"
                },
                {
                    "criteria": {"column": "position_source_type_id", "operator_type": "isnotnull"},
                    "condition_for_next_rule": "And"
                },
                {
                    "criteria": {"column": "value", "operator_type": ">=", "value": 100}
                }
            ]
        },
        [0, 4, 5]  # holding AND id not null AND value >= 100
    ):
        passed += 1
    else:
        failed += 1
    
    # TEST 19: Deeply nested criteria within single rule
    if run_test(
        "Deep nesting: OR of ANDs of ORs",
        base_df,
        {
            "rules": [{
                "criteria": {
                    "or": [
                        {
                            "and": [
                                {"or": [
                                    {"column": "value", "operator_type": "=", "value": 100},
                                    {"column": "value", "operator_type": "=", "value": 310}
                                ]},
                                {"column": "category", "operator_type": "=", "value": "holding"}
                            ]
                        },
                        {
                            "and": [
                                {"column": "position_source_type_id", "operator_type": "=", "value": 26},
                                {"column": "value", "operator_type": "<", "value": 200}
                            ]
                        }
                    ]
                }
            }]
        },
        [0, 3, 5]  # ((value=100 OR value=310) AND holding) OR (id=26 AND value<200)
                   # Index 0: (100 OR false) AND holding = TRUE
                   # Index 3: (false) OR (26 AND 180<200) = TRUE  
                   # Index 5: (false OR 310) AND holding = TRUE
    ):
        passed += 1
    else:
        failed += 1
    
    # TEST 20: Case sensitivity test for string operators
    mixed_case_df = pl.DataFrame({
        "name": ["John", "JOHN", "john", "Jane", "JANE", "jane"],
        "category": ["holding"] * 6
    })
    
    if run_test(
        "Case insensitive equality",
        mixed_case_df,
        {
            "rules": [{
                "criteria": {"column": "name", "operator_type": "=", "value": "JOHN"}
            }]
        },
        [0, 1, 2]  # All variations of "john" match
    ):
        passed += 1
    else:
        failed += 1
    
    # TEST 21: Between with equal bounds
    if run_test(
        "BETWEEN with equal lower and upper bounds",
        base_df,
        {
            "rules": [{
                "criteria": {
                    "column": "value",
                    "operator_type": "between", 
                    "value": [200, 200]
                }
            }]
        },
        [1]  # Only value=200 matches
    ):
        passed += 1
    else:
        failed += 1
    
    # TEST 22: Multiple inequality operators chained
    if run_test(
        "Multiple inequalities: >100 AND <=250 AND !=200",
        base_df,
        {
            "rules": [
                {
                    "criteria": {"column": "value", "operator_type": ">", "value": 100},
                    "condition_for_next_rule": "And"
                },
                {
                    "criteria": {"column": "value", "operator_type": "<=", "value": 250},
                    "condition_for_next_rule": "And"
                },
                {
                    "criteria": {"column": "value", "operator_type": "!=", "value": 200}
                }
            ]
        },
        [3, 4]  # 180 and 250 (not 200)
    ):
        passed += 1
    else:
        failed += 1
    
    # TEST 23: Test with all nulls
    null_only_df = pl.DataFrame({
        "x": [None, None, None],
        "category": ["holding", "reference", "holding"]
    })
    
    if run_test(
        "All NULL values with != operator",
        null_only_df,
        {
            "rules": [{
                "criteria": {"column": "x", "operator_type": "!=", "value": 5}
            }]
        },
        []  # NULL != 5 returns False for all
    ):
        passed += 1
    else:
        failed += 1
    
    # TEST 24: IN operator with mixed types (string representation)
    if run_test(
        "IN operator with string list representation",
        base_df,
        {
            "rules": [{
                "criteria": {
                    "column": "position_source_type_id",
                    "operator_type": "in",
                    "value": "9, 10, 15"  # Without brackets
                }
            }]
        },
        [0, 1, 4, 5]  # IDs 9, 10, 15
    ):
        passed += 1
    else:
        failed += 1
    
    # TEST 25: Verify order matters - same operators different order
    if run_test(
        "Order verification: (A OR B) AND C != A OR (B AND C)",
        base_df,
        {
            "rules": [
                {
                    "criteria": {"column": "value", "operator_type": "=", "value": 100},
                    "condition_for_next_rule": "Or"
                },
                {
                    "criteria": {"column": "value", "operator_type": "=", "value": 310},
                    "condition_for_next_rule": "And"
                },
                {
                    "criteria": {"column": "category", "operator_type": "=", "value": "holding"}
                }
            ]
        },
        [0, 5]  # (value=100 OR value=310) AND holding
                # Index 0: (TRUE OR FALSE) AND TRUE = TRUE
                # Index 5: (FALSE OR TRUE) AND TRUE = TRUE
    ):
        passed += 1
    else:
        failed += 1
    
    # Final summary
    print("\n" + "="*70)
    print("TEST SUMMARY")
    print("="*70)
    print(f"Total tests: {passed + failed}")
    print(f"âœ… Passed: {passed}")
    print(f"âŒ Failed: {failed}")
    
    if failed == 0:
        print("\nğŸ‰ SUCCESS! All tests passed - Polars implementation matches OO exactly!")
    else:
        print(f"\nâš ï¸ WARNING: {failed} test(s) failed - implementations don't match!")
    
    # Note: The implementations now match exactly with the following behavior:
    # 1. NULL handling: NULL values return False for all operators except isnull/isnotnull
    # 2. LIKE/NOTLIKE: Case-insensitive substring matching
    # 3. IN/NOTIN: Handle both single values and lists, with proper NULL behavior
    # 4. BETWEEN/NOTBETWEEN: Inclusive range checking
    # 5. apply_to: Filters by category (holding/reference/both)
    # 6. condition_for_next_rule: Chains rules with AND/OR logic
    # 7. Empty rules: All rows pass (return True for all)
```

```python
def generate_random_data(n_rows: int) -> pl.DataFrame:
    """Generate a large DataFrame with random data"""
    random.seed(42)  # For reproducibility
    
    # Generate random data
    data = {
        "id": list(range(n_rows)),
        "value1": [random.randint(1, 1000) for _ in range(n_rows)],
        "value2": [random.uniform(0, 100) for _ in range(n_rows)],
        "value3": [random.randint(-50, 50) for _ in range(n_rows)],
        "category": [random.choice(["holding", "reference"]) for _ in range(n_rows)],
        "status": [random.choice(["active", "inactive", "pending"]) for _ in range(n_rows)],
        "score": [random.uniform(-1, 1) for _ in range(n_rows)],
        "type_id": [random.choice([1, 2, 3, 4, 5, None]) for _ in range(n_rows)],
        "amount": [random.randint(100, 10000) if random.random() > 0.1 else None for _ in range(n_rows)],
        "flag": [random.choice([True, False, None]) for _ in range(n_rows)],
        "code": [''.join(random.choices(string.ascii_uppercase, k=3)) for _ in range(n_rows)],
        "ratio": [random.uniform(0, 2) for _ in range(n_rows)],
        "count": [random.randint(0, 100) for _ in range(n_rows)],
        "temperature": [random.uniform(-20, 40) for _ in range(n_rows)],
        "priority": [random.choice([1, 2, 3, 4, 5]) for _ in range(n_rows)]
    }
    
    return pl.DataFrame(data)


def time_execution(func, *args, **kwargs):
    """Time the execution of a function"""
    start = time.perf_counter()
    result = func(*args, **kwargs)
    end = time.perf_counter()
    return end - start, result


def run_oo_evaluation(df: pl.DataFrame, rules_json: Dict[str, Any]) -> tuple:
    """Run OO evaluation and return time and results"""
    rows = df.to_dicts()
    start = time.perf_counter()
    results = CriteriaEvaluator.evaluate_rows(rules_json, rows)
    end = time.perf_counter()
    return end - start, results


def run_polars_evaluation(df: pl.DataFrame, rules_json: Dict[str, Any]) -> tuple:
    """Run Polars evaluation and return time and results"""
    rules = rules_json.get("rules", [])
    start = time.perf_counter()
    expr = PolarsExpressionBuilder.build_rules_expr(rules)
    polars_result_series = df.select(expr.alias("result"))["result"]
    
    # Handle scalar result
    if len(polars_result_series) == 1 and len(df) > 1:
        results = [polars_result_series[0]] * len(df)
    else:
        results = polars_result_series.to_list()
    
    end = time.perf_counter()
    return end - start, results


def run_speed_test(test_name: str, df: pl.DataFrame, rules_json: Dict[str, Any], warmup: bool = True):
    """Run a single speed test comparing OO and Polars"""
    print(f"\n{'='*70}")
    print(f"TEST: {test_name}")
    print(f"Dataset size: {len(df):,} rows Ã— {len(df.columns)} columns")
    print(f"Number of rules: {len(rules_json.get('rules', []))}")
    print('='*70)
    
    if warmup:
        # Warmup runs to avoid cold start effects
        run_oo_evaluation(df.head(100), rules_json)
        run_polars_evaluation(df.head(100), rules_json)
    
    # OO Evaluation
    oo_time, oo_results = run_oo_evaluation(df, rules_json)
    print(f"OO Time:     {oo_time*1000:.2f}ms")
    
    # Polars Evaluation
    polars_time, polars_results = run_polars_evaluation(df, rules_json)
    print(f"Polars Time: {polars_time*1000:.2f}ms")
    
    # Verify results match
    matches = oo_results == polars_results
    print(f"Results match: {'âœ… Yes' if matches else 'âŒ No'}")
    
    # Calculate speedup
    speedup = oo_time / polars_time if polars_time > 0 else float('inf')
    print(f"Speedup: {speedup:.1f}x faster with Polars")
    
    # Count matches
    match_count = sum(oo_results)
    print(f"Rows matching criteria: {match_count:,} ({match_count/len(df)*100:.1f}%)")
    
    return oo_time, polars_time, matches


# ======================================================================
# MAIN SPEED TESTS
# ======================================================================

if __name__ == "__main__":
    print("\n" + "="*70)
    print("PERSPECTIVE RULES ENGINE - PERFORMANCE COMPARISON")
    print("OO Implementation vs Polars Implementation")
    print("="*70)
    
    # Test with different dataset sizes
    sizes = [1_000, 10_000, 100_000, 500_000, 1_000_000]
    
    # Store results for summary
    all_results = []
    
    # Test 1: Simple equality filter
    print("\n\nğŸ”¥ TEST SUITE 1: Simple Equality Filter")
    print("-" * 70)
    simple_rules = {
        "rules": [{
            "criteria": {
                "column": "value1",
                "operator_type": ">",
                "value": 500
            }
        }]
    }
    
    for size in sizes:
        df = generate_random_data(size)
        oo_time, polars_time, matches = run_speed_test(
            f"Simple filter ({size:,} rows)",
            df,
            simple_rules
        )
        all_results.append((size, "Simple", oo_time, polars_time))
    
    # Test 2: Complex AND/OR chains
    print("\n\nğŸ”¥ TEST SUITE 2: Complex AND/OR Chains")
    print("-" * 70)
    complex_rules = {
        "rules": [
            {
                "criteria": {"column": "value1", "operator_type": ">", "value": 300},
                "condition_for_next_rule": "And"
            },
            {
                "criteria": {"column": "value2", "operator_type": "<", "value": 50},
                "condition_for_next_rule": "Or"
            },
            {
                "criteria": {"column": "category", "operator_type": "=", "value": "holding"},
                "condition_for_next_rule": "And"
            },
            {
                "criteria": {"column": "amount", "operator_type": "isnotnull"},
                "condition_for_next_rule": "And"
            },
            {
                "criteria": {"column": "priority", "operator_type": "in", "value": "[1, 2, 3]"}
            }
        ]
    }
    
    for size in [1_000, 10_000, 100_000]:  # Smaller sizes for complex rules
        df = generate_random_data(size)
        oo_time, polars_time, matches = run_speed_test(
            f"Complex AND/OR chains ({size:,} rows)",
            df,
            complex_rules
        )
        all_results.append((size, "Complex", oo_time, polars_time))
    
    # Test 3: Deeply nested criteria
    print("\n\nğŸ”¥ TEST SUITE 3: Deeply Nested Criteria")
    print("-" * 70)
    nested_rules = {
        "rules": [{
            "criteria": {
                "or": [
                    {
                        "and": [
                            {"column": "value1", "operator_type": "between", "value": [200, 800]},
                            {"column": "category", "operator_type": "=", "value": "holding"}
                        ]
                    },
                    {
                        "and": [
                            {"column": "amount", "operator_type": ">", "value": 5000},
                            {"column": "status", "operator_type": "in", "value": ["active", "pending"]}
                        ]
                    },
                    {
                        "or": [
                            {"column": "flag", "operator_type": "=", "value": True},
                            {"column": "priority", "operator_type": "=", "value": 1}
                        ]
                    }
                ]
            }
        }]
    }
    
    for size in [1_000, 10_000, 100_000]:
        df = generate_random_data(size)
        oo_time, polars_time, matches = run_speed_test(
            f"Deeply nested criteria ({size:,} rows)",
            df,
            nested_rules
        )
        all_results.append((size, "Nested", oo_time, polars_time))
    
    # Test 4: String operations (LIKE)
    print("\n\nğŸ”¥ TEST SUITE 4: String Operations")
    print("-" * 70)
    string_rules = {
        "rules": [
            {
                "criteria": {"column": "code", "operator_type": "like", "value": "A"},
                "condition_for_next_rule": "And"
            },
            {
                "criteria": {"column": "status", "operator_type": "!=", "value": "inactive"}
            }
        ]
    }
    
    for size in [1_000, 10_000, 50_000]:
        df = generate_random_data(size)
        oo_time, polars_time, matches = run_speed_test(
            f"String operations ({size:,} rows)",
            df,
            string_rules
        )
        all_results.append((size, "String", oo_time, polars_time))
    
    # Test 5: Many rules (10+ rules)
    print("\n\nğŸ”¥ TEST SUITE 5: Many Rules (Stress Test)")
    print("-" * 70)
    many_rules = {
        "rules": [
            {"criteria": {"column": "value1", "operator_type": ">", "value": 100}, "condition_for_next_rule": "And"},
            {"criteria": {"column": "value1", "operator_type": "<", "value": 900}, "condition_for_next_rule": "And"},
            {"criteria": {"column": "value2", "operator_type": ">=", "value": 10}, "condition_for_next_rule": "And"},
            {"criteria": {"column": "value2", "operator_type": "<=", "value": 90}, "condition_for_next_rule": "And"},
            {"criteria": {"column": "value3", "operator_type": "!=", "value": 0}, "condition_for_next_rule": "And"},
            {"criteria": {"column": "category", "operator_type": "=", "value": "holding"}, "condition_for_next_rule": "Or"},
            {"criteria": {"column": "status", "operator_type": "=", "value": "active"}, "condition_for_next_rule": "And"},
            {"criteria": {"column": "amount", "operator_type": "isnotnull"}, "condition_for_next_rule": "And"},
            {"criteria": {"column": "flag", "operator_type": "!=", "value": False}, "condition_for_next_rule": "And"},
            {"criteria": {"column": "priority", "operator_type": "in", "value": "[1, 2, 3, 4]"}, "condition_for_next_rule": "And"},
            {"criteria": {"column": "ratio", "operator_type": "between", "value": [0.5, 1.5]}, "condition_for_next_rule": "And"},
            {"criteria": {"column": "count", "operator_type": ">", "value": 10}, "condition_for_next_rule": "And"},
            {"criteria": {"column": "temperature", "operator_type": "notbetween", "value": [-10, -5]}, "condition_for_next_rule": "And"},
            {"criteria": {"column": "code", "operator_type": "notlike", "value": "Z"}, "condition_for_next_rule": "And"},
            {"criteria": {"column": "type_id", "operator_type": "notin", "value": "[4, 5]"}}
        ]
    }
    
    for size in [1_000, 10_000, 50_000]:
        df = generate_random_data(size)
        oo_time, polars_time, matches = run_speed_test(
            f"15 rules stress test ({size:,} rows)",
            df,
            many_rules
        )
        all_results.append((size, "Many Rules", oo_time, polars_time))
    
    # ======================================================================
    # SUMMARY
    # ======================================================================
    
    print("\n\n" + "="*70)
    print("PERFORMANCE SUMMARY")
    print("="*70)
    
    print("\nğŸ“Š Results Table:")
    print(f"{'Dataset Size':>12} | {'Test Type':15} | {'OO Time (ms)':>12} | {'Polars (ms)':>12} | {'Speedup':>8}")
    print("-" * 75)
    
    for size, test_type, oo_time, polars_time in all_results:
        speedup = oo_time / polars_time if polars_time > 0 else float('inf')
        print(f"{size:12,} | {test_type:15} | {oo_time*1000:12.2f} | {polars_time*1000:12.2f} | {speedup:7.1f}x")
    
    # Calculate average speedup by test type
    print("\nğŸ“ˆ Average Speedup by Test Type:")
    test_types = {}
    for _, test_type, oo_time, polars_time in all_results:
        if test_type not in test_types:
            test_types[test_type] = []
        speedup = oo_time / polars_time if polars_time > 0 else float('inf')
        test_types[test_type].append(speedup)
    
    for test_type, speedups in test_types.items():
        avg_speedup = sum(speedups) / len(speedups)
        print(f"  {test_type:15}: {avg_speedup:6.1f}x faster")
    
    # Overall average
    all_speedups = []
    for _, _, oo_time, polars_time in all_results:
        speedup = oo_time / polars_time if polars_time > 0 else float('inf')
        all_speedups.append(speedup)
    
    overall_avg = sum(all_speedups) / len(all_speedups)
    print(f"\nğŸš€ Overall Average Speedup: {overall_avg:.1f}x")
    
    # Performance scaling analysis
    print("\nğŸ“‰ Performance Scaling (1k to 1M rows):")
    simple_results = [(size, oo, pl) for size, test, oo, pl in all_results if test == "Simple"]
    if len(simple_results) >= 2:
        small = simple_results[0]  # 1k rows
        large = simple_results[-1]  # largest dataset
        
        oo_scale = (large[1] / small[1]) / (large[0] / small[0])
        polars_scale = (large[2] / small[2]) / (large[0] / small[0])
        
        print(f"  OO scaling factor: {oo_scale:.2f} (1.0 = perfect linear)")
        print(f"  Polars scaling factor: {polars_scale:.2f} (1.0 = perfect linear)")
        
        if polars_scale < oo_scale:
            print(f"  âœ… Polars scales {oo_scale/polars_scale:.1f}x better than OO!")
```

```python
import polars as pl
import json

def parse_perspective_data(json_data: dict) -> pl.DataFrame:
    """
    Parses the Perspective nested JSON structure into a single flat Polars DataFrame.
    
    Args:
        json_data: The dictionary containing the full JSON response
        
    Returns:
        pl.DataFrame: Unified dataframe containing holdings, lookthroughs, and references
    """
    
    dfs = []
    
    # Helper function to process a specific section (e.g., holding.positions)
    def process_section(parent_key, sub_key, category_name, type_name):
        # Navigate safely to the data
        section_data = json_data.get(parent_key, {}).get(sub_key)
        
        if not section_data:
            return None

        # Convert the Dictionary { "id": {data} } to a List [ {id: "id", ...data} ]
        # This is much faster for DataFrame creation than iterating rows later
        flattened_rows = []
        for id_key, attributes in section_data.items():
            row = attributes.copy()
            row['node_id'] = id_key  # Preserve the key (e.g., "160046", "80504")
            flattened_rows.append(row)
            
        if not flattened_rows:
            return None

        # Create DataFrame
        df = pl.DataFrame(flattened_rows)
        
        # Add structural metadata columns
        df = df.with_columns([
            pl.lit(category_name).alias("category"), # 'holding' or 'reference'
            pl.lit(type_name).alias("record_type")   # 'position' or 'lookthrough'
        ])
        
        return df

    # 1. Extract Holding Positions
    df_hold_pos = process_section("holding", "positions", "holding", "position")
    if df_hold_pos is not None: dfs.append(df_hold_pos)

    # 2. Extract Holding Lookthroughs
    df_hold_look = process_section("holding", "lookthroughs", "holding", "lookthrough")
    if df_hold_look is not None: dfs.append(df_hold_look)

    # 3. Extract Reference Positions
    df_ref_pos = process_section("reference", "positions", "reference", "position")
    if df_ref_pos is not None: dfs.append(df_ref_pos)

    # 4. Combine everything
    if not dfs:
        return pl.DataFrame()

    # 'how="diagonal"' creates a union of all columns. 
    # If 'parent_instrument_id' exists in lookthroughs but not positions, 
    # positions get nulls automatically.
    final_df = pl.concat(dfs, how="diagonal")

    # 5. Clean up types (Optional but recommended)
    # Ensure numeric IDs are actually numeric if they look like integers
    if "parent_instrument_id" in final_df.columns:
        final_df = final_df.with_columns(
            pl.col("parent_instrument_id").cast(pl.Int64, strict=False)
        )

    return final_df

# ==========================================
# Example Usage with your data structure
# ==========================================
if __name__ == "__main__":
    # Mocking the data from your screenshots
    raw_json = {
        "perspective_configurations": {}, # (omitted for brevity)
        "holding": {
            "positions": {
                "160046": { "instrument_identifier": 126892, "weight": 1.0, "is_class_position": False },
                "160047": { "instrument_identifier": 126892, "weight": 1.0, "is_class_position": False }
            },
            "lookthroughs": {
                "80504": { "instrument_identifier": 426291, "weight": 0.035, "parent_instrument_id": 518186 },
                "80505": { "instrument_identifier": 426291, "weight": 0.035, "parent_instrument_id": 518186 }
            }
        },
        "reference": {
            "positions": {
                "160044": { "instrument_identifier": 518186, "weight": 1.0, "is_class_position": False }
            }
        }
    }

    df = parse_perspective_data(raw_json)
    
    print("Shape:", df.shape)
    print(df)
```

```python
import polars as pl
import json
from typing import Any, Dict, List, Optional, Union

# ======================================================================
# 1. RULE ENGINE (Optimized Polars Implementation)
# ======================================================================

class PolarsExpressionBuilder:
    """
    Optimized Polars Expression Builder.
    Focuses on Expression Graph minimization and ensuring Common Subexpression Elimination (CSE) 
    can occur effectively within the Polars optimizer.
    """

    @staticmethod
    def normalize_value(val: Any, op: str) -> Any:
        """Pre-process value in Python to save work in Polars execution."""
        if op in ("in", "notin", "between", "notbetween") and isinstance(val, str):
            # Parse stringified lists: "[1, 2]" -> [1, 2]
            parts = val.strip("[]").split(",")
            normalized = []
            for x in parts:
                x = x.strip()
                if x.isdigit():
                    normalized.append(int(x))
                else:
                    normalized.append(x.lower() if isinstance(x, str) else x)
            return normalized
        elif isinstance(val, str):
            return val.lower()
        return val

    @staticmethod
    def _get_col_expr(col_name: str, as_string: bool = False) -> pl.Expr:
        """
        Generates a standardized column expression.
        Using standard nodes helps Polars CSE (Common Subexpression Elimination).
        """
        c = pl.col(col_name)
        if as_string:
            return c.cast(pl.Utf8).str.to_lowercase()
        return c

    @staticmethod
    def build_leaf_expr(criteria: Dict[str, Any]) -> pl.Expr:
        """Build optimized leaf expression."""
        col_name = criteria["column"].lower()
        op = criteria.get("operator_type", "").lower()
        raw_val = criteria.get("value")
        
        # 1. Handle Null Checks (Fast path)
        if op == "isnull":
            return pl.col(col_name).is_null()
        if op == "isnotnull":
            return pl.col(col_name).is_not_null()

        # 2. Normalize value
        val = PolarsExpressionBuilder.normalize_value(raw_val, op)

        # 3. Determine if we need string comparison logic
        is_str_op = isinstance(val, str) or (isinstance(val, list) and any(isinstance(x, str) for x in val))
        col = PolarsExpressionBuilder._get_col_expr(col_name, as_string=is_str_op)

        # 4. Build Operator Expression
        expr = pl.lit(False) # Default fallback

        if op == "=":
            expr = col.eq(val)
        elif op == "!=":
            expr = col.ne(val)
        elif op == ">":
            expr = col.gt(val)
        elif op == ">=":
            expr = col.ge(val)
        elif op == "<":
            expr = col.lt(val)
        elif op == "<=":
            expr = col.le(val)
        
        elif op == "in":
            if isinstance(val, list):
                str_vals = [v for v in val if isinstance(v, str)]
                num_vals = [v for v in val if not isinstance(v, str)]
                
                conditions = []
                if str_vals:
                    conditions.append(PolarsExpressionBuilder._get_col_expr(col_name, True).is_in(str_vals))
                if num_vals:
                    conditions.append(pl.col(col_name).is_in(num_vals))
                
                if conditions:
                    expr = conditions[0]
                    for c in conditions[1:]:
                        expr = expr | c
            else:
                expr = col.eq(val)

        elif op == "notin":
            if isinstance(val, list):
                str_vals = [v for v in val if isinstance(v, str)]
                num_vals = [v for v in val if not isinstance(v, str)]
                
                expr = pl.lit(True)
                if str_vals:
                    expr = expr & PolarsExpressionBuilder._get_col_expr(col_name, True).is_in(str_vals).not_()
                if num_vals:
                    expr = expr & pl.col(col_name).is_in(num_vals).not_()
            else:
                expr = col.ne(val)

        elif op == "between":
            if isinstance(val, list) and len(val) >= 2:
                expr = col.ge(val[0]) & col.le(val[1])
        
        elif op == "notbetween":
            if isinstance(val, list) and len(val) >= 2:
                expr = (col.lt(val[0]) | col.gt(val[1]))
            else:
                expr = pl.lit(True)

        elif op == "like":
            if isinstance(val, str):
                expr = col.str.contains(val, literal=True) 

        elif op == "notlike":
            if isinstance(val, str):
                expr = col.str.contains(val, literal=True).not_()
            else:
                expr = pl.lit(True)

        return expr.fill_null(False)

    @staticmethod
    def build_criteria_expr(criteria: Dict[str, Any]) -> pl.Expr:
        """Recursively build boolean expression tree."""
        if "and" in criteria:
            exprs = [PolarsExpressionBuilder.build_criteria_expr(c) for c in criteria["and"]]
            if not exprs: return pl.lit(True)
            final = exprs[0]
            for e in exprs[1:]: final = final & e
            return final
            
        elif "or" in criteria:
            exprs = [PolarsExpressionBuilder.build_criteria_expr(c) for c in criteria["or"]]
            if not exprs: return pl.lit(False)
            final = exprs[0]
            for e in exprs[1:]: final = final | e
            return final
            
        return PolarsExpressionBuilder.build_leaf_expr(criteria)

    @staticmethod
    def build_rules_expr(rules: List[Dict[str, Any]]) -> pl.Expr:
        """Builds the master expression for the rule list."""
        if not rules:
            return pl.lit(True)

        combined_expr = None

        for i, rule in enumerate(rules):
            criteria = rule.get("criteria", {})
            if isinstance(criteria, str):
                criteria = json.loads(criteria)
            
            current_expr = PolarsExpressionBuilder.build_criteria_expr(criteria)

            # Optimizing 'apply_to'
            apply_to = (rule.get("apply_to") or "both").lower()
            if apply_to == "holding":
                current_expr = current_expr & pl.col("category").eq("holding")
            elif apply_to == "reference":
                current_expr = current_expr & pl.col("category").eq("reference")

            if i == 0:
                combined_expr = current_expr
            else:
                prev_rule = rules[i-1]
                condition = prev_rule.get("condition_for_next_rule", "And").lower()
                if condition == "or":
                    combined_expr = combined_expr | current_expr
                else:
                    combined_expr = combined_expr & current_expr

        return combined_expr if combined_expr is not None else pl.lit(False)

# ======================================================================
# 2. PARSER & PROCESSING LOGIC
# ======================================================================

def parse_perspective_data(json_data: dict) -> pl.DataFrame:
    """Parses the Perspective nested JSON structure into a single flat Polars DataFrame."""
    dfs = []
    
    def process_section(parent_key, sub_key, category_name, type_name):
        section_data = json_data.get(parent_key, {}).get(sub_key)
        if not section_data: return None

        flattened_rows = []
        for id_key, attributes in section_data.items():
            row = attributes.copy()
            row['node_id'] = id_key
            flattened_rows.append(row)
            
        if not flattened_rows: return None

        df = pl.DataFrame(flattened_rows)
        df = df.with_columns([
            pl.lit(category_name).alias("category"),
            pl.lit(type_name).alias("record_type")
        ])
        return df

    # Extract sections
    sections = [
        ("holding", "positions", "holding", "position"),
        ("holding", "lookthroughs", "holding", "lookthrough"),
        ("reference", "positions", "reference", "position")
    ]
    
    for p_key, s_key, cat, rec_type in sections:
        d = process_section(p_key, s_key, cat, rec_type)
        if d is not None: dfs.append(d)

    if not dfs: return pl.DataFrame()

    final_df = pl.concat(dfs, how="diagonal")

    # Type cleanup
    if "parent_instrument_id" in final_df.columns:
        final_df = final_df.with_columns(pl.col("parent_instrument_id").cast(pl.Int64, strict=False))
    
    return final_df

def apply_perspective_rules(df: pl.DataFrame, rules_json: Dict[str, Any]) -> pl.DataFrame:
    """Applies the dynamic rules to filter the DataFrame."""
    rules = rules_json.get("rules", [])
    if not rules:
        return df
        
    filter_expr = PolarsExpressionBuilder.build_rules_expr(rules)
    return df.filter(filter_expr)

def rescale_lookthrough_weights(df: pl.DataFrame) -> pl.DataFrame:
    """
    Rescales lookthrough weights so that they sum up to their parent position's weight.
    This is typically used after filtering to ensure portfolio consistency.
    """
    # 1. Identify Parent Weights (from Positions)
    # We assume 'instrument_identifier' in a Position record is the ID referenced by 'parent_instrument_id'
    if "parent_instrument_id" not in df.columns:
        return df

    parents = df.filter(
        pl.col("record_type") == "position"
    ).select([
        pl.col("instrument_identifier").alias("parent_instrument_id"),
        pl.col("weight").alias("parent_target_weight")
    ])

    # 2. Calculate sum of CURRENT lookthrough weights per parent
    lt_sums = df.filter(
        pl.col("record_type") == "lookthrough"
    ).group_by("parent_instrument_id").agg(
        pl.col("weight").sum().alias("current_lt_sum")
    )

    # 3. Calculate Scaling Factors
    # Join parents and sums to calculate factor: parent_target / current_sum
    scaling_factors = parents.join(lt_sums, on="parent_instrument_id", how="inner")
    scaling_factors = scaling_factors.with_columns(
        (pl.col("parent_target_weight") / pl.col("current_lt_sum")).alias("scale_factor")
    )
    
    # 4. Join factors back to main DF
    df_with_factors = df.join(
        scaling_factors.select(["parent_instrument_id", "scale_factor"]), 
        on="parent_instrument_id", 
        how="left"
    )
    
    # 5. Apply scaling only to lookthroughs where we have a valid factor
    final_df = df_with_factors.with_columns(
        pl.when(
            (pl.col("record_type") == "lookthrough") & (pl.col("scale_factor").is_not_null())
        )
        .then(pl.col("weight") * pl.col("scale_factor"))
        .otherwise(pl.col("weight"))
        .alias("weight")
    ).drop("scale_factor")
    
    return final_df

# ==========================================
# Example Usage
# ==========================================
if __name__ == "__main__":
    # 1. Mock Data
    raw_json = {
        "holding": {
            "positions": {
                "160046": { "instrument_identifier": 126892, "weight": 1.0 }, # Parent
                "160047": { "instrument_identifier": 999999, "weight": 0.5 }  # Unrelated Position
            },
            "lookthroughs": {
                # Sum = 0.7 (0.3 + 0.4). Should scale to 1.0 (Parent 160046/126892)
                "80504": { "instrument_identifier": 426291, "weight": 0.3, "parent_instrument_id": 126892 },
                "80505": { "instrument_identifier": 426292, "weight": 0.4, "parent_instrument_id": 126892 },
                # This one we will filter out to demonstrate rescaling
                "80506": { "instrument_identifier": 666666, "weight": 0.1, "parent_instrument_id": 126892 } 
            }
        }
    }

    # 2. Parse
    df = parse_perspective_data(raw_json)
    print("--- Original DataFrame ---")
    print(df.select(["node_id", "record_type", "weight", "parent_instrument_id"]))

    # 3. Define Rules (Example: Exclude instrument 666666)
    rules_json = {
        "rules": [{
            "criteria": {
                "column": "instrument_identifier",
                "operator_type": "!=",
                "value": 666666
            }
        }]
    }

    # 4. Apply Rules
    filtered_df = apply_perspective_rules(df, rules_json)
    print("\n--- Filtered DataFrame (80506 removed) ---")
    print(filtered_df.select(["node_id", "record_type", "weight", "parent_instrument_id"]))

    # 5. Rescale Lookthroughs
    # Remaining LT weights: 0.3 + 0.4 = 0.7. Parent target = 1.0.
    # Factor should be 1.0 / 0.7 ~= 1.428
    rescaled_df = rescale_lookthrough_weights(filtered_df)
    print("\n--- Rescaled DataFrame ---")
    print(rescaled_df.select(["node_id", "record_type", "weight", "parent_instrument_id"]))
```
